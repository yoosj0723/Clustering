```python
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm


# 경고 메시지 무시 설정
warnings.filterwarnings('ignore')

# 폰트 경로 설정
font_path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'

# 폰트 설정
fontprop = fm.FontProperties(fname=font_path)

# 상관계수 행렬을 계산합니다.
correlation_matrix = encoding_data.corr()

# 히트맵을 그립니다.
plt.figure(figsize=(20, 20))
ax = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', annot_kws={'fontsize': 5})
ax.set_title('상관계수 히트맵', fontproperties=fontprop)  # 폰트 적용

# x축, y축 라벨에 한글 폰트 적용
ax.set_xticklabels(ax.get_xticklabels(), fontproperties=fontprop)
ax.set_yticklabels(ax.get_yticklabels(), fontproperties=fontprop)

plt.show()
```

```python
import pandas as pd

# 상관계수가 특정 기준을 넘는 변수들을 선택합니다.
threshold = 0.63  # 상관계수 기준 설정
high_correlation = correlation_matrix[(abs(correlation_matrix) > threshold) & (abs(correlation_matrix) < 1)]

# 상관계수가 높은 변수들의 변수명과 값을 출력합니다.
high_correlation = high_correlation.dropna(how='all', axis=1).dropna(how='all', axis=0)

# 변수명과 값을 출력합니다.
for col in high_correlation.columns:
    variables = high_correlation[high_correlation[col].notnull()].index.tolist()
    values = high_correlation[col][high_correlation[col].notnull()].tolist()

    for variable, value in zip(variables, values):
        print(f"{col} - {variable}: {value}")
```

# 다중공선성

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

# 독립변수들로 구성된 데이터프레임을 준비합니다.
X = encoding_data  # 독립변수 데이터프레임

# 상수항을 추가합니다.
X = sm.add_constant(X)

# VIF 계산을 위한 데이터프레임을 생성합니다.
vif_data = pd.DataFrame()
vif_data["Variable"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

# 결과를 데이터프레임으로 변환
vif_data_df = pd.DataFrame(vif_data)

# 결과를 출력
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(vif_data_df)
```

```python
#전략기획 및 전개, 성과관리
#인적자원 관리, 인적자원 개발
#분석 및 설계, 구현 및 테스트
#품질 보증 및 형상 관리, 측정 및 분석
#문제 해결 및 리스크 관리, 프로세스 개선 관리

# 삭제할 이름 리스트
columns_to_delete = ["전략기획 및 전개", "성과관리", "인적자원 관리", "인적자원 개발", "분석 및 설계", "구현 및 테스트",
                     "품질 보증 및 형상 관리", "측정 및 분석", "문제 해결 및 리스크 관리", "프로세스 개선 관리"]
# A열과 B열을 합친 평균 값을 계산하여 C열에 추가
encoding_data['전략기획 및 전개_성과관리'] = (encoding_data['전략기획 및 전개'] + encoding_data['성과관리']) / 2
encoding_data['인적자원 관리_인적자원 개발'] = (encoding_data['인적자원 관리'] + encoding_data['인적자원 개발']) / 2
encoding_data['분석 및 설계_구현 및 테스트'] = (encoding_data['분석 및 설계'] + encoding_data['구현 및 테스트']) / 2
encoding_data['품질 보증 및 형상 관리_측정 및 분석'] = (encoding_data['품질 보증 및 형상 관리'] + encoding_data['측정 및 분석']) / 2
encoding_data['문제 해결 및 리스크 관리_프로세스 개선 관리'] = (encoding_data['문제 해결 및 리스크 관리'] + encoding_data['프로세스 개선 관리']) / 2
# 데이터프레임에서 특정 열들을 제거
encoding_data2 = encoding_data.drop(columns=columns_to_delete)
encoding_data2
```

# PCA

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# 데이터 정규화 (StandardScaler 사용)
#scaler = StandardScaler()
#data_scaled = scaler.fit_transform(data)

# PCA 모델 생성 및 주성분 추출
pca = PCA(n_components=2)  # 주성분 개수를 지정
X_pca = pca.fit_transform(encoding_data)

# 주성분 결과 확인
explained_variance_ratio = pca.explained_variance_ratio_  # 주성분의 설명된 분산 비율
print("주성분의 설명된 분산 비율:", explained_variance_ratio)

# 주성분으로 이루어진 데이터 프레임 생성
principal_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])

# 주성분 데이터와 원본 데이터 병합
result_df = pd.concat([principal_df, pd.DataFrame(encoding_data)], axis=1)

# 주성분 분석 결과 출력
print("주성분 분석 결과:")
print(result_df.head())

# 주성분의 계수(요소적재값) 확인
component_loading = pd.DataFrame(pca.components_.T, columns=['PC1', 'PC2'], index=encoding_data.columns)
print("주성분의 계수:")
print(component_loading)

# 가장 많은 영향을 주는 변수 확인
most_influential_variable_pc1 = component_loading['PC1'].idxmax()
most_influential_variable_pc2 = component_loading['PC2'].idxmax()
print("PC1에서 가장 많은 영향을 주는 변수:", most_influential_variable_pc1)
print("PC2에서 가장 많은 영향을 주는 변수:", most_influential_variable_pc2)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# 클러스터 개수 설정 (예시: 4)
K = 2

# 데이터 정규화 (StandardScaler 사용)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_pca)

# K-means 클러스터링 모델 생성 및 학습
kmeans = KMeans(n_clusters=K)
kmeans.fit(X_pca)

# 각 데이터 포인트의 클러스터 할당 결과 확인
cluster_labels = kmeans.labels_

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 차원 축소된 결과와 클러스터 레이블을 DataFrame으로 생성
df_pca = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
df_pca['Cluster'] = cluster_labels

# Seaborn의 "Set3" 팔레트에서 색상 가져오기
colors = sns.color_palette('Set3', K)

for cluster in range(K):
    cluster_data = df_pca[df_pca['Cluster'] == cluster]
    plt.scatter(cluster_data['PC1'], cluster_data['PC2'], color=colors[cluster], label=f'Cluster {cluster}')

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score
from sklearn.metrics import silhouette_samples
import numpy as np

# 클러스터링 결과에 대한 실루엣 계수 계산
silhouette_avg = silhouette_score(X_scaled, cluster_labels)
print(f"전체 데이터의 실루엣 평균 계수: {silhouette_avg}")

# 각 데이터 포인트의 실루엣 계수 계산
sample_silhouette_values = silhouette_samples(X_scaled, cluster_labels)

# 각 클러스터의 평균 실루엣 계수 계산
for i in range(K):
    cluster_silhouette_avg = np.mean(sample_silhouette_values[cluster_labels == i])
    print(f"클러스터 {i}의 평균 실루엣 계수: {cluster_silhouette_avg}")
```

```python
import matplotlib.pyplot as plt

# 실루엣 다이어그램 그리기
fig, ax = plt.subplots()
y_lower = 10
for i in range(K):
    # 클러스터 i의 실루엣 계수 정렬
    ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
    ith_cluster_silhouette_values.sort()

    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i

    # 클러스터 i에 대한 실루엣 다이어그램 그리기
    color = plt.cm.get_cmap("Spectral")(float(i) / K)
    ax.fill_betweenx(np.arange(y_lower, y_upper),
                     0, ith_cluster_silhouette_values,
                     facecolor=color, edgecolor=color, alpha=0.7)

    # 클러스터 i의 레이블 텍스트 추가
    ax.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

    # 다음 클러스터의 실루엣 다이어그램을 그리기 위해 y_lower 업데이트
    y_lower = y_upper + 10

# 실루엣 평균 계수의 수평선 그리기
ax.axvline(x=silhouette_avg, color="red", linestyle="--")

plt.title("Silhouette Diagram")
plt.xlabel("Silhouette coefficient values")
plt.ylabel("Cluster label")

# 실루엣 다이어그램 출력
plt.show()
```

```python
import matplotlib.pyplot as plt

# 실루엣 다이어그램 그리기
fig, ax = plt.subplots()
y_lower = 10
for i in range(K):
    # 클러스터 i의 실루엣 계수 정렬
    ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
    ith_cluster_silhouette_values.sort()

    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i

    # 클러스터 i에 대한 실루엣 다이어그램 그리기
    color = plt.cm.get_cmap("Spectral")(float(i) / K)
    ax.fill_betweenx(np.arange(y_lower, y_upper),
                     0, ith_cluster_silhouette_values,
                     facecolor=color, edgecolor=color, alpha=0.7)

    # 클러스터 i의 레이블 텍스트 추가
    ax.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

    # 다음 클러스터의 실루엣 다이어그램을 그리기 위해 y_lower 업데이트
    y_lower = y_upper + 10

# 실루엣 평균 계수의 수평선 그리기
ax.axvline(x=silhouette_avg, color="red", linestyle="--")

plt.title("Silhouette Diagram")
plt.xlabel("Silhouette coefficient values")
plt.ylabel("Cluster label")

# 실루엣 다이어그램 출력
plt.show()
```

# 새 섹션

```python
import numpy as np
from scipy.cluster import hierarchy
import matplotlib.pyplot as plt

# 데이터 정규화 (StandardScaler 사용)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_pca)

# 덴드로그램 클러스터링
distance_matrix = hierarchy.distance.pdist(X_scaled)  # 거리 행렬 계산
linkage_matrix = hierarchy.linkage(distance_matrix, method='average')  # 연결 행렬 계산

plt.figure(figsize=(12, 6))
dendrogram = hierarchy.dendrogram(linkage_matrix)  # 덴드로그램 생성

# 덴드로그램 시각화
plt.xlabel('Data Index')
plt.ylabel('Distance')
plt.title('Dendrogram')
plt.show()
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score

# 클러스터링 결과
labels = hierarchy.fcluster(linkage_matrix, t=3, criterion='maxclust')  # 적절한 클러스터 수인 t 값을 지정해야 합니다.

# 실루엣 계수 계산
silhouette_vals = silhouette_samples(X_scaled, labels)

# 전체 데이터의 실루엣 점수 평균
silhouette_avg = silhouette_score(X_scaled, labels)
print(f"전체 데이터의 실루엣 점수 평균: {silhouette_avg}")

# 각 클러스터별 실루엣 점수 계산 및 출력
unique_labels = np.unique(labels)
for label in unique_labels:
    cluster_silhouette_vals = silhouette_vals[labels == label]
    cluster_silhouette_avg = np.mean(cluster_silhouette_vals)
    print(f"클러스터 {label}의 실루엣 점수: {cluster_silhouette_avg}")
```
